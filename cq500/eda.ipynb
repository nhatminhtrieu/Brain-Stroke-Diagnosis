```python
import os
import pandas as pd

# Define the base path for the 'Unknown Study' folder
unknown_study_path = 'archive/CQ500CT0/Unknown Study/'

# Define the output CSV file path
output_csv_path = 'output.csv'

# Function to find the folder with the smallest number of DICOM files and save to CSV
def find_folder_with_fewest_dicom_files_and_save_to_csv(base_path, output_csv):
    # Initialize variables to track the folder with the fewest DICOM files
    min_dicom_count = float('inf')  # Start with infinity
    target_folder_name = None

    # Iterate through all folders in the base path
    for folder_name in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder_name)
        
        # Check if it is a directory
        if os.path.isdir(folder_path):
            # Count DICOM files in the directory
            dicom_files = [f for f in os.listdir(folder_path) if f.endswith('.dcm')]
            dicom_count = len(dicom_files)

            # If this folder has fewer DICOM files than previously found, update target
            if dicom_count > 0 and dicom_count < min_dicom_count:
                min_dicom_count = dicom_count
                target_folder_name = folder_name
                print(f'Found {dicom_count} DICOM files in: {folder_name}')

    # Prepare data for CSV if a valid target folder was found
    if target_folder_name:
        data = [{'name': 'CQ500-CT-0', 'Source folder': target_folder_name}]
        df = pd.DataFrame(data)
        
        # Save DataFrame to CSV
        df.to_csv(output_csv, index=False)
        print(f'Data saved to {output_csv} for folder: {target_folder_name}')
    else:
        print('No folders with DICOM files found.')

# Execute the function
# find_folder_with_fewest_dicom_files_and_save_to_csv(unknown_study_path, output_csv_path)
```

```python
import os
import pandas as pd

# Define paths
filtered_reads_path = 'filtered_reads.csv'

# Function to find the folder with the smallest number of DICOM files, ignoring folders with 'bone' in their names
def find_folder_with_fewest_dicom_files(base_path):
    min_dicom_count = float('inf')
    target_folder_name = None

    if os.path.exists(base_path):
        for folder_name in os.listdir(base_path):
            folder_path = os.path.join(base_path, folder_name)

            if (os.path.isdir(folder_path)
                and 'bone' not in folder_name.lower()
                and 'helical' not in folder_name.lower()
                and 'hires' not in folder_name.lower()
                and 'hi res' not in folder_name.lower()):

                # # Prioritize 'plain' folders
                if 'plain' in folder_name.lower() and 'thin' not in folder_name.lower():
                    return folder_name  # Immediate return on first 'plain' folder

                dicom_files = [f for f in os.listdir(folder_path) if f.endswith('.dcm')]
                dicom_count = len(dicom_files)

                if dicom_count > 0 and dicom_count < min_dicom_count:
                    min_dicom_count = dicom_count
                    target_folder_name = folder_name
    else:
        print(f'Path does not exist: {base_path}')
        return None

    return target_folder_name

# Read filtered_reads.csv into a DataFrame
df = pd.read_csv(filtered_reads_path)

# Iterate over each row in the DataFrame to process CQ500-CT-427 and others
for index, row in df.iterrows():
    row_name = row['name']

    # Replace '-' with '' 
    row_name = row_name.replace('-', '')
    
    # Constructing a new path based on 'CQ500-CT-427'
    specific_folder_path = os.path.join('archive', row_name, 'Unknown Study')

    # Find the folder with the fewest DICOM files in that specific directory
    found_folder_name = find_folder_with_fewest_dicom_files(specific_folder_path)

    # Add found folder name to a new column in the DataFrame
    if found_folder_name:
        df.loc[index, 'Source Folder'] = found_folder_name
        print(f'Found and added Source Folder: {found_folder_name} for {row_name}')
    else:
        print(f'No DICOM folders found for {row_name}.')

# Save updated DataFrame back to CSV
df.to_csv(filtered_reads_path, index=False)
print(f'Updated {filtered_reads_path} with new Source Folder information.')
```

```python
import pandas as pd
import os

src_dir = 'archive'

# Count the number of dicom files in each folder
def count_dicom_files(folder_path):
    if not os.path.exists(folder_path):
        return 0
    dicom_files = [f for f in os.listdir(folder_path) if f.endswith('.dcm')]
    return len(dicom_files)

# Create a DataFrame to store the results
results = []

# Iterate based on the csv file
df = pd.read_csv('filtered_reads.csv')

for index, row in df.iterrows():
    folder_name = row['name']
    source_folder = str(row['Source Folder'])
    folder_path = os.path.join(src_dir, folder_name, 'Unknown Study', source_folder)

    # Count the number of dicom files in the folder
    dicom_count = count_dicom_files(folder_path)

    # Append the results to the list
    results.append({'name': folder_name, 'dicom_count': dicom_count})

# Create a new DataFrame from the results
results_df = pd.DataFrame(results)

# Save the results to a new CSV file
results_df.to_csv('dicom_counts.csv', index=False)

# Print min and max dicom counts
min_count = results_df['dicom_count'].min()
max_count = results_df['dicom_count'].max()

print(f'Minimum DICOM count: {min_count}')
print(f'Maximum DICOM count: {max_count}')

# Print row that 'Source Folder' is empty
empty_source_folder = df[df['Source Folder'].isnull()]
print(empty_source_folder)

# Delete row that 'Source Folder' is empty
df = df.dropna(subset=['Source Folder'])
df.to_csv('filtered_reads_without_nan.csv', index=False)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('filtered_reads_without_nan.csv')

postfixes = ['ICH', 'IPH', 'IVH', 'SDH', 'EDH', 'SAH', 'BleedLocation-Left', 'BleedLocation-Right', 'ChronicBleed', 'Fracture',
             'CalvarialFracture', 'OtherFracture', 'MassEffect', 'MidlineShift']
prefixes = ['R1:', 'R2:', 'R3:']

# Create patient-level columns using majority voting (2 radiologists agree)
for postfix in postfixes:
    # Get all columns for this hemorrhage type
    cols = [f'{prefix}{postfix}' for prefix in prefixes]

    # Create new column with majority agreement
    df[f'patient_{postfix}'] = (df[cols].sum(axis=1) >= 2).astype(int)

# Create the 'patient_any' column if any patient_* column is 1
patient_cols = [f'patient_{p}' for p in postfixes]
df['patient_labels'] = (df[patient_cols].sum(axis=1) > 0).astype(int)

# Drop rows with empty Source Folder
df = df.dropna(subset=['Source Folder'])

# Save to new CSV
df.to_csv('cq500.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit
import numpy as np

TEST_SIZE = 0.15
VALID_SIZE = 0.25

def split_dataset_for_multilabel(patient_scan_labels, test_size=0.15, val_size=0.25, random_state=42):
    # Extract the labels from the DataFrame
    prefix = 'patient_'
    postfix = ['ICH', 'IPH', 'IVH', 'SDH', 'EDH', 'SAH', 'BleedLocation-Left', 'BleedLocation-Right', 'ChronicBleed', 'Fracture',
               'CalvarialFracture', 'OtherFracture', 'MassEffect', 'MidlineShift']
    labels = patient_scan_labels[[f'{prefix}{p}' for p in postfix]].values

    if test_size > 0:
        # First split: train + test
        msss = MultilabelStratifiedShuffleSplit(n_splits=1, test_size=test_size, random_state=random_state)
        train_idx, test_idx = next(msss.split(patient_scan_labels, labels))

        train_labels = patient_scan_labels.iloc[train_idx]
        test_labels = patient_scan_labels.iloc[test_idx]

        # Second split: train + validation
        msss_val = MultilabelStratifiedShuffleSplit(n_splits=1, test_size=val_size, random_state=random_state)
        train_idx, val_idx = next(msss_val.split(train_labels, labels[train_idx]))

        train_labels_final = train_labels.iloc[train_idx]
        val_labels = train_labels.iloc[val_idx]

    else:
        # Only split into train and validation if test_size is 0
        msss_val = MultilabelStratifiedShuffleSplit(n_splits=1, test_size=val_size, random_state=random_state)
        train_idx, val_idx = next(msss_val.split(patient_scan_labels, labels))

        train_labels_final = patient_scan_labels.iloc[train_idx]
        val_labels = patient_scan_labels.iloc[val_idx]
        test_labels = None

    return train_labels_final, val_labels, test_labels
```

```python
patient_scan_labels = pd.read_csv('cq500.csv')

train_labels, val_labels, test_labels = split_dataset_for_multilabel(patient_scan_labels, test_size=TEST_SIZE)

# Define new columns for hemorrhage types (these should be in your original DataFrame)
prefix = 'patient_'
postfix = ['ICH', 'IPH', 'IVH', 'SDH', 'EDH', 'SAH', 'BleedLocation-Left', 'BleedLocation-Right', 'ChronicBleed', 'Fracture',
           'CalvarialFracture', 'OtherFracture', 'MassEffect', 'MidlineShift']

hemorrhage_columns = [f'{prefix}{p}' for p in postfix]

def count_hemorrhage_types(data):
    # Original counts for hemorrhage types
    counts = {col: data[col].sum() for col in hemorrhage_columns}
    unique_patients = {col: data[data[col] == 1].shape[0] for col in hemorrhage_columns}

    no_hem_count = (data[hemorrhage_columns].sum(axis=1) == 0).sum()

    counts['No Hemorrhage'] = no_hem_count
    unique_patients['No Hemorrhage'] = no_hem_count

    return counts, unique_patients

# Count for training set
train_counts, train_unique_patients = count_hemorrhage_types(train_labels)

# Count for validation set
val_counts, val_unique_patients = count_hemorrhage_types(val_labels)

# Count for test set
test_counts, test_unique_patients = count_hemorrhage_types(test_labels)
```

```python
# Update summary data with new category
summary_data = {
    'Hemorrhage Type': hemorrhage_columns + ['No Hemorrhage'],
    'Train Counts': [train_counts[col] for col in hemorrhage_columns] + [train_counts['No Hemorrhage']],
    'Validation Counts': [val_counts[col] for col in hemorrhage_columns] + [val_counts['No Hemorrhage']],
    'Test Counts': [test_counts[col] for col in hemorrhage_columns] + [test_counts['No Hemorrhage']]
}

# Create a DataFrame for the summary table
summary_table = pd.DataFrame(summary_data)

# Display the summary table
summary_table
```
